### 5. Случайная величина распределена равномерно на отрезке $[\theta, 2\theta]$


#### f)
Сгенерируйте выборку объема n = 100 для некоторого значения параметра $\theta$. Вычислите указанные выше доверительные интервалы для доверительной вероятности 0.95

#### g)

#### h)

```python
theta = 1

import numpy as np
import random

def p(x: float) -> float:
    return 1 / theta if theta <= x <= 2 * theta else 0

def F(x: float) -> float:
    if (x <= theta):
        return 0
    elif x >= 2 * theta:
        return 1
    else:
        return (x / theta - 1)

def F_inv(y: float) -> float:
    return theta * (y + 1)
    
n = 100
arr : list = [float(F_inv(random.random())) for _ in range(n)]

arr
```

### Точный метод:

$$
P(\frac{max(x_i)}{\sqrt[n]{\frac{1 + \beta}{2}} + 1} < \theta < \frac{max(x_i)}{\sqrt[n]{\frac{1 - \beta}{2}} + 1}) = \beta
$$

```python

beta = 0.95
left_part = np.max(arr) / (((1 + beta) / 2) ** (1 / n) + 1)
right_part = np.max(arr) / (((1 - beta) / 2) ** (1 / n) + 1)
print(f'Доверительный интервал: ({left_part}, {right_part})')
print(f'Длина интервала: {right_part - left_part}')
```

### Асимптотический метод:

$$
P(\tilde{\theta}_1 - \frac{t_2 \cdot 2S\sqrt{n - 1}}{3n} < \theta < \tilde{\theta}_1 - \frac{t_1 \cdot 2S\sqrt{n - 1}}{3n}) = \beta
$$

```python
t_1 = -1.96
t_2 = 1.96

theta_1 = 2 / 3 * np.mean(arr)
s = np.sqrt(np.sum([(x - np.mean(arr)) ** 2 for x in arr]) / (n - 1))

left_part = theta_1 - (t_2 * 2 * s * np.sqrt(n - 1)) / (3 * n)
right_part = theta_1 - (t_1 * 2 * s * np.sqrt(n - 1)) / (3 * n)
print(f'Доверительный интервал: ({left_part}, {right_part})')
print(f'Длина интервала: {right_part - left_part}')
```

#### g)
Численно постройте бутстраповский доверительный интервал.

```python
bootstrap_iteration = 1000

theta_wave = (n + 1) / (2 * n + 1) * np.max(arr) # Оценка ОМП

bootstrap_delta = []

for i in range(bootstrap_iteration):
    bootstrap_arr = np.random.choice(arr, size=len(arr), replace=True)
    bootstrap_theta = (n + 1) / (2 * n + 1) * np.max(bootstrap_arr)
    bootstrap_delta.append(bootstrap_theta - theta_wave)

bootstrap_delta.sort()
t_1 = bootstrap_delta[int(bootstrap_iteration * (1 - beta) / 2)]
t_2 = bootstrap_delta[int(bootstrap_iteration * (1 + beta) / 2)]
right_part = -(t_1 - theta_wave)
left_part = -(t_2 - theta_wave)
print(f'Доверительный интервал: ({left_part}, {right_part})')
print(f'Длина интервала: {right_part - left_part}')
```
